```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
from google.colab import drive
drive.mount('/content/drive') # Change the mount point to '/content/drive'
```

```python
image=Image.open("/content/drive/MyDrive/n/100002.jpg")
plt.imshow(image)
```

```python
img_graysale=image.convert('L')
plt.imshow(img_graysale,cmap='gray')
print(img_graysale.size,img_graysale.mode,img_graysale.format)
```

```python

transpose1=image.transpose(Image.FLIP_LEFT_RIGHT)
plt.subplot(4,2,1)
plt.imshow(transpose1)
plt.title("left to right")

transpose2=image.transpose(Image.FLIP_TOP_BOTTOM)
plt.subplot(4,2,2)
plt.imshow(transpose2)
plt.title("top to bottom")

transpose3=image.transpose(Image.ROTATE_90)
plt.subplot(4,2,3)
plt.imshow(transpose3)
plt.title("rotate 90")

transpose_ll=image.transpose(Image.TRANSPOSE)
plt.subplot(4,2,4)
plt.imshow(transpose_ll)
plt.title("transpose")
```

```python
newsize=(1100,400)
image_resize=image.resize(newsize,Image.BICUBIC)
plt.imshow(image_resize)
plt.title("resize bibi")
```

```python
plt.imshow(img_graysale)
img_quantize=image.quantize(1)
plt.imshow(img_quantize)
```

```python
from PIL import ImageEnhance
img_contrast = ImageEnhance.Contrast(image)
img_C = img_contrast.enhance(2)
plt.title("ontrast enhanement")
plt.imshow(img_C)
```
