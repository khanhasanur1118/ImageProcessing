```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

from google.colab import drive
drive.mount('/content/drive')
```

```python
single=Image.open('/content/drive/MyDrive/image/11111.jpg')
dope=Image.open('/content/drive/MyDrive/image/22222.jpg')
selfi=Image.open('/content/drive/MyDrive/image/44444.jpg')
plt.subplot(1,1,1)
plt.imshow(single)
print(single.size,single.mode,single.format_description)
plt.subplot(2,2,2)
plt.imshow(dope)
print(dope.size,dope.mode,dope.format_description)
plt.subplot(1,4,2)
plt.imshow(selfi)
print(selfi.size,selfi.mode,selfi.format_description)

```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
plt.subplots_adjust(wspace=0.5)

axes[0].imshow(single)
axes[0].axis('on')
axes[0].set_title("single photo")
axes[1].imshow(dope)
axes[1].axis('on')
axes[1].set_title("dope photo")
axes[2].imshow(selfi)
axes[2].axis('off')
axes[2].set_title("selfi photo")
plt.show()
```

```python
box = (280, 80, 1200, 1000)
single_crop = single.crop(box)
plt.imshow(single_crop)


```

```python
from PIL import ImageFont
from PIL import ImageDraw
import matplotlib.font_manager
watermarked_image=single_crop.copy()
draw=ImageDraw.Draw(watermarked_image)
font_path=matplotlib.font_manager.findSystemFonts(fontpaths=None, fontext='ttf')[0]
font=ImageFont.truetype(font_path,50)
draw.text((200,60),"Hasanur Roshid Khan",fill=(255,0,255),font=font)
plt.imshow(watermarked_image)
```

```python
from PIL import ImageEnhance
img_contrast = ImageEnhance.Contrast(single_crop)
enh_img=img_contrast.enhance(1.4)
enh_img.show("80% more contrast")
plt.title(" enhanement")
plt.imshow(enh_img)
```
